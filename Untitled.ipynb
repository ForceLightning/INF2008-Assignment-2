# Data Loading

```python
data = []
labels = []
```

```python
for i, classification in enumerate(["negative", "positive"]):
    print(1 * "---", classification)
    for root, dirs, files in os.walk(f"./numpy_2d/{classification}"):
        for file in files:
            print(2 * "---", file)
            arr = np.load(f"./numpy_2d/{classification}/{file}")
            data.append(arr)
            labels.append(i)
```

```python
filtered_freqs = shifted_freqs.copy()
# filtered_freqs[np.abs(filtered_freqs) < 10] = 0
# filtered_freqs[np.abs(filtered_freqs) > 40] = 0
filtered_freqs[:20] = 0
filtered_freqs[90:] = 0
filtered_freqs[:,:96] = 0
filtered_freqs[:,264:] = 0
# filtered_freqs[53:55, 174:185] = 0
```

```python
sns.heatmap(np.abs(filtered_freqs))
```

```python
reconstructed = np.fft.ifft2(filtered_freqs)
sns.heatmap(np.abs(reconstructed))
```

```python
def fft_crude_low_pass(frame, xmin, xmax, ymin, ymax):
    frame[:ymin] = 0
    frame[ymax:] = 0
    frame[:,:xmin] = 0
    frame[:,xmax:] = 0
```

```python
fft_data = []
for i, frame in enumerate(data):
    freqs = np.fft.fft2(frame)
    freqs = np.fft.fftshift(freqs)
    freqs = fft_crude_low_pass(freqs, 96, 264, 20, 90)
    reconstructed = np.fft.ifft2(filtered_freqs)
    reconstructed = np.abs(reconstructed)
    fft_data.append(reconstructed)
```
